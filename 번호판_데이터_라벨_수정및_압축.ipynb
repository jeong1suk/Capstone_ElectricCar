```python
import os
from glob import glob 

txt_list = sorted(glob('/content/train/labels/*.txt'))
for fname in txt_list:
  pass

val_txt_list = sorted(glob('/content/valid/labels/*.txt'))
for fname in val_txt_list:
  pass

for i in range(len(txt_list)):
  with open(txt_list[i],'r') as f:
    line = f.readlines()
    
  line = line[0].split()
  line[0] = '15'

  result = ' '.join(s for s in line)
  print(result)

                  
  with open(txt_list[i],'w') as f:
    f.write(result)
      
```
